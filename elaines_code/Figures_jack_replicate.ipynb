```python
# load data
filepath5 = "out_J_0.5_K_0.5"
filepath6 = "out_J_0.3_K_0.2"
filepath7 = "out_J_1_K_0.2"
data5 = np.loadtxt(filepath5)
data6 = np.loadtxt(filepath6)
data7 = np.loadtxt(filepath7)
```

```python
def split_data(d):

    t = []
    x = []
    y = []
    theta = []

    for row in d:
        t.append([row[0]])
        x.append(row[1:d.shape[1]:3])
        y.append(row[2:d.shape[1]:3])
        theta.append(row[3:d.shape[1]:3])

    return np.array(t),np.array(x),np.array(y),theta
```

```python
def f(theta):
    return 0.45*(1.+np.cos(theta))

def theta_rgb(theta):
    rgb=[]
    for th in theta:
        rgb.append([f(th), f(th-2.*np.pi/3.), f(th+2.*np.pi/3.)])
    return np.array(rgb)
```

```python
t1,x1,y1,theta1 = split_data(data5)
t2,x2,y2,theta2 = split_data(data6)
t3,x3,y3,theta3 = split_data(data7)
title = ['$J=0.5, K=0.5$', '$J=0.3, K=-0.2$', '$J=1, K=-0.2$']
```

```python
fig,ax = plt.subplots(5,3, figsize=(20,30))

for j in range(5):
    ax[j][0].scatter(x1[j], y1[j], color=theta_rgb(theta1[j]))
    ax[j][1].scatter(x2[j], y2[j], color=theta_rgb(theta2[j]))
    ax[j][2].scatter(x3[j], y3[j], color=theta_rgb(theta3[j]))
    ax[j][0].set_xlabel('x', fontsize=14)
    ax[j][1].set_xlabel('x', fontsize=14)
    ax[j][2].set_xlabel('x', fontsize=14)
    ax[j][0].tick_params(axis='both', which='major', labelsize=14)
    ax[j][1].tick_params(axis='both', which='major', labelsize=14)
    ax[j][2].tick_params(axis='both', which='major', labelsize=14)
    ax[j][0].set_ylim([-1.5,1.5])
    ax[j][1].set_ylim([-1.5,1.5])
    ax[j][2].set_ylim([-1.5,1.5])
    ax[j][0].set_xlim([-1.5,1.5])
    ax[j][1].set_xlim([-1.5,1.5])
    ax[j][2].set_xlim([-1.5,1.5])
    if j==0:
        ax[j][0].set_title(title[0], fontsize=14)
        ax[j][1].set_title(title[1], fontsize=14)
        ax[j][2].set_title(title[2], fontsize=14)
    ax[j][0].set_ylabel('Snapshot at t={}\ny'.format(int(t1[j])), fontsize=14)

plt.show()
fig.savefig('RANDOM_p5a.png')
```
